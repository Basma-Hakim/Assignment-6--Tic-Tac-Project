```python
import random
theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

board_keys = []
computerMove=' '
NewMove=[]
TowMoves=[]
computerTowMoves=[]



for key in theBoard:
    board_keys.append(key)

''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''

def printBoard(board):
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])

def secMove(list1,i):
    if list1[i]== 1:
        return random.choice([4,2,5,9,3,7])
    elif list1[i]==2:
        return random.choice([1,3,5,8])
    elif list1[i]==3:
        return random.choice([1,2,5,7,6,9])
    elif list1[i]==4:
        return random.choice([5,6,1,7])
    elif list1[i]==5:
        return random.choice([1,2,3,4,6,7,8,9])
    elif list1[i]==6:
        return random.choice([4,5,9,3])
    elif list1[i]==7:
        return random.choice([1,4,5,3,8,9])
    elif list1[i]==8:
        return random.choice([7,9,5,2])
    elif list1[i]==9:
        return random.choice([7,8,5,1,6,3])
    
    
    
def NextMove(list1,i,j):
           
            
            if (list1[i],list1[j]) in [(2,3),(3,2),(7,4) ,(4,7),(9,5) ,(5,9)]:
                return 1
            
            elif (list1[i],list1[j]) in [(6,3),(3,6),(7,8) ,(8,7),(5,1),(1,5)]:  
                return 9
            
            elif (list1[i],list1[j]) in [(3,5),(5,3),(9,8) ,(8,9),(1,4),(4,1)]:
                return 7
            
            elif (list1[i],list1[j]) in [(7,5) ,(5,7),(9,6) ,(6,9),(1,2),(2,1)]:
                return 3    
                
            elif (list1[i],list1[j]) in [(4,5) ,(5,4),(9,3),(3,9)]:
                return 6  
                
            elif (list1[i],list1[j]) in [(6,5) ,(5,6), (1,7),(7,1)]:
                return 4
               
            elif (list1[i],list1[j]) in [(2,5) ,(5,2),(7,9),(9,7)]:
                return 8    
                
            elif (list1[i],list1[j]) in [(5,8) ,(8,5),(1,3),(3,1)]:
                return 2
            
            elif (list1[i],list1[j]) in [(1,9),(9,1),(2,8),(8,2),(3,7),(7,3), (4,6),(6,4)]:
                return 5
            
            else:
                return None
                

# Now we'll write the main function which has all the gameplay functionality.
def game():
    possibleMoves=[1,2,3,4,5,6,7,8,9]
    turn = 'X'
    count = 0
    computerMoves=0
    flag=''
    compMove1=''
    compMove2=''
    computerMove=' '
    NewMove=[]
    Comp1Moves=[]
    Comp2Moves=[]
    theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

    board_keys = []
    
    for i in range(6):
        
       
                
                
                printBoard(theBoard)
                print("All possible moves are: ", possibleMoves )
                print("O moves sequence: ", Comp1Moves)
                print("X moves sequence: ", Comp2Moves)
                
                if i==0:
                    print ("i = ",i)
                    turn='O'
                    compMove1 = random.choice(possibleMoves)
                    theBoard[str(compMove1)] = turn
                    possibleMoves.remove(compMove1)      
                    Comp1Moves.append(compMove1)
                    count += 1

            


                    turn='X'
                    compMove2 = random.choice(possibleMoves)
                    theBoard[str(compMove2)] = turn
                    possibleMoves.remove(compMove2)      
                    Comp2Moves.append(compMove2)
                    count += 1


                
                
               # NewMove=possibleMoves

        
          
                   
                    
                        
                if i==1:
                    print ("i = ",i)
                    turn='O'
                    compMove1 = secMove(Comp1Moves,0)
                    while compMove1 not in possibleMoves:
                        compMove1 = secMove(Comp1Moves,0)
                    print ("computer1 move is, ",  compMove1)

                         
                         
                    count += 1   
                    theBoard[str(compMove1)] = turn 
                    Comp1Moves.append(compMove1)
                    possibleMoves.remove(compMove1)
                 
                
                    turn='X'
                    compMove2= secMove(Comp2Moves,0)
                    while compMove2 not in possibleMoves:
                        compMove2 = secMove(Comp2Moves,0)
                    print ("computer1 move is, ",  compMove2)
                    
                    
                                             
                    count += 1   
                    theBoard[str(compMove2)] = turn 
                    Comp2Moves.append(compMove2)
                    possibleMoves.remove(compMove2)
                        
                # this part for computers to win          
                if i==2 and possibleMoves!=[]:
                    print ("i = ",i)
                    turn='O'
                    compMove1 = NextMove(Comp1Moves,0,1)
                    
                    if compMove1==None or compMove1 not in possibleMoves  :
                        compMove1 = NextMove(Comp2Moves,0,1)
                        if compMove1==None or compMove1 not in possibleMoves  :
                            compMove1= secMove(Comp1Moves,1)
                            while compMove1 not in possibleMoves  :
                                compMove1= secMove(Comp1Moves,1)
                            else:
                                compMove1=random.choice(possibleMoves)
    
                
                    count += 1   
                    theBoard[str(compMove1)] = turn 
                    Comp1Moves.append(compMove1)
                    possibleMoves.remove(compMove1)
                    
                    turn='X'
                    
                    compMove2 = NextMove(Comp2Moves,0,1)
                    if compMove2==None or compMove2 not in possibleMoves  :
                        compMove2 = NextMove(Comp1Moves,0,1)
                        if compMove2==None or compMove2 not in possibleMoves  :
                            compMove2= secMove(Comp2Moves,1)
                            while compMove2 not in possibleMoves  :
                                compMove2= secMove(Comp2Moves,1)
                            else:
                                compMove2=random.choice(possibleMoves)
                   
                                
                    count += 1   
                    theBoard[str(compMove2)] = turn 
                    Comp2Moves.append(compMove2)
                    possibleMoves.remove(compMove2)
                    
                    
                if i==3 and possibleMoves!=[]:
                    print ("i = ",i)
                    turn='O' 
                    flag='No'
                    for x in range(3):
                        if flag=='No':
                            for y in range(x+1,3):
                                compMove1 = NextMove(Comp1Moves,x,y)
                                if compMove1 in possibleMoves  :
                                    flag='yes'
                                    break
                        else:
                            break
                            
                    if compMove1==None:
                        flag='No'                       
                        for x in range(3):
                            if flag=='No':
                                for y in range(x+1,3):
                                    compMove1 = NextMove(Comp2Moves,x,y)
                                    if compMove1 in possibleMoves  :
                                        flag='yes'
                                        break
                            else:
                                break
                                
                    if compMove1==None:
                        compMove1= secMove(Comp1Moves,2)
                        while compMove1 not in possibleMoves  :
                            compMove1= secMove(Comp1Moves,2)
                        else:
                            compMove1=random.choice(possibleMoves)

                    count += 1   
                    theBoard[str(compMove1)] = turn 
                    Comp1Moves.append(compMove1)
                    possibleMoves.remove(compMove1)

                    turn='X' 
                    flag='No'
                    for x in range(3):
                        if flag=='No':
                            for y in range(x+1,3):
                                compMove2 = NextMove(Comp2Moves,x,y)
                                if compMove2 in possibleMoves  :
                                    flag='yes'
                                    break
                        else:
                            break

                    if compMove2==None:
                        flag='No'
                        for x in range(2):
                            if flag=='No':
                                for y in range(x+1,3):
                                    compMove2 = NextMove(Comp1Moves,x,y)
                                    if compMove2 in possibleMoves  :
                                        flag='yes'
                                        break
                            else:
                                break

                        if compMove2==None:
                            compMove2= secMove(Comp2Moves,2)
                            while compMove2 not in possibleMoves  :
                                compMove2= secMove(Comp2Moves,2)
                            else:
                                compMove2=random.choice(possibleMoves)

                    count += 1   
                    theBoard[str(compMove2)] = turn 
                    Comp2Moves.append(compMove2)
                    possibleMoves.remove(compMove2)



        # Now we will check if player X or O has won,for every move after 5 moves. 
                if count >= 5 or possibleMoves ==[]:
                    if theBoard['7'] == theBoard['8'] == theBoard['9'] != ' ': # across the top
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['7'] + " won. ****")                
                        break
                    elif theBoard['4'] == theBoard['5'] == theBoard['6'] != ' ': # across the middle
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['4'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ': # across the bottom
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['1'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['4'] == theBoard['7'] != ' ': # down the left side
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['1'] + " won. ****")
                        break
                    elif theBoard['2'] == theBoard['5'] == theBoard['8'] != ' ': # down the middle
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['2'] + " won. ****")
                        break
                    elif theBoard['3'] == theBoard['6'] == theBoard['9'] != ' ': # down the right side
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['3'] + " won. ****")
                        break 
                    elif theBoard['7'] == theBoard['5'] == theBoard['3'] != ' ': # diagonal
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['7'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['5'] == theBoard['9'] != ' ': # diagonal
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['1'] + " won. ****")
                        break 

                # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
                    if count == 9:
                        print("\nGame Over.\n")                
                        print("It's a Tie!!")
                        printBoard(theBoard)
                        break

                      
       

    # Now we will ask if player wants to restart the game or not.
    
    restart = input("Do want to play Again?(y/n)")
    if restart == "y" or restart == "Y":  
        
        for key in board_keys:
            theBoard[key] = " "
        

        game()

if __name__ == "__main__":
    game() 
```
