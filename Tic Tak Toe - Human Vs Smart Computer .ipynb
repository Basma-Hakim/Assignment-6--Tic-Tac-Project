```python
 import random
theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

board_keys = []
computerMove=' '
NewMove=[]
TowMoves=[]
computerTowMoves=[]



for key in theBoard:
    board_keys.append(key)

''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''

def printBoard(board):
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])


def NextMove(list1,i,j):
           
            
            if (list1[i],list1[j]) in [(2,3),(3,2),(7,4) ,(4,7),(9,5) ,(5,9)]:
                return 1
            
            elif (list1[i],list1[j]) in [(6,3),(3,6),(7,8) ,(8,7),(5,1),(1,5)]:  
                return 9
            
            elif (list1[i],list1[j]) in [(3,5),(5,3),(9,8) ,(8,9),(1,4),(4,1)]:
                return 7
            
            elif (list1[i],list1[j]) in [(7,5) ,(5,7),(9,6) ,(6,9),(1,2),(2,1)]:
                return 3    
                
            elif (list1[i],list1[j]) in [(4,5) ,(5,4),(9,3),(3,9)]:
                return 6  
                
            elif (list1[i],list1[j]) in [(6,5) ,(5,6), (1,7),(7,1)]:
                return 4
               
            elif (list1[i],list1[j]) in [(2,5) ,(5,2),(7,9),(9,7)]:
                return 8    
                
            elif (list1[i],list1[j]) in [(5,8) ,(8,5),(1,3),(3,1)]:
                return 2
            
            elif (list1[i],list1[j]) in [(1,9),(9,1),(2,8),(8,2),(3,7),(7,3), (4,6),(6,4)]:
                return 5
            
            else:
                return None
                

# Now we'll write the main function which has all the gameplay functionality.
def game():
    possibleMoves=[1,2,3,4,5,6,7,8,9]
    turn = 'X'
    count = 0
    computerMoves=0
    
    computerMove=' '
    NewMove=[]
    TowMoves=[]
    computerTowMoves=[]
    theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

    board_keys = []
    
    for i in range(6):
        
       
                turn='O'
                
                printBoard(theBoard)
                print("All possible moves are: ", possibleMoves )
                print("O moves sequence: ", TowMoves)
                print("X moves sequence: ", computerTowMoves)
                move = input("It's your turn," + turn + " , move to which place? ")

                      
                
                while  int(move) not in possibleMoves:
                    print("That place is not one of possible moves:", possibleMoves,"\n move to which place?")
                    move = input()
                   
               
                
                else:
                    
                    theBoard[move] = turn
                    
                    TowMoves.append(int(move))
                    possibleMoves.remove(int(move))
                   
                    count += 1

            
                turn='X'


                
                
                NewMove=possibleMoves

        
                
                if i==0:
                    computerMove=random.choice(NewMove)
                  
                    
                    possibleMoves.remove(computerMove)
                    
                    theBoard[str(computerMove)] = turn
                    computerTowMoves.append(computerMove)
                    count += 1
                   
                    
                        
                if i==1:
                    computerMove= NextMove(TowMoves,0,1)
                    
                    
                        
                    if computerMove==None or computerMove not in possibleMoves:
                        
                        computerMove=random.choice(possibleMoves)
                        
                         
                    count += 1   
                    theBoard[str(computerMove)] = turn 
                    computerTowMoves.append(computerMove)
                    possibleMoves.remove(computerMove)
                 
                        
               # this part for computer to win          
                if i>=2 and possibleMoves!=[]:
                        
                    computerMove = NextMove(computerTowMoves,i-2,i-1)
                    if computerMove==None or computerMove not in possibleMoves  :
                        if i>=3 :
                            computerMove = NextMove(computerTowMoves,0,2)
                            if computerMove==None or computerMove not in possibleMoves :
                                
                                computerMove= NextMove(TowMoves,i-1,i)
                                
                                if computerMove==None or computerMove not in possibleMoves  :
                                    
                                    computerMove= NextMove(TowMoves,i,0)
                                    
                                    if computerMove==None or computerMove not in possibleMoves  :
                                        computerMove=random.choice(possibleMoves)
                                        
                                
                        else:
                           
                            computerMove= NextMove(TowMoves,i-1,i)
                           
                            if computerMove==None or computerMove not in possibleMoves  :
                                
                                computerMove= NextMove(TowMoves,i,0)
                                
                                if computerMove==None or computerMove not in possibleMoves  :
                                    computerMove=random.choice(possibleMoves)
                                    
                        
                                
                            
                
                    count += 1   
                    theBoard[str(computerMove)] = turn 
                    computerTowMoves.append(computerMove)
                    possibleMoves.remove(computerMove)
                   
                      
                    
        # Now we will check if player X or O has won,for every move after 5 moves. 
                if count >= 5 or possibleMoves ==[]:
                    if theBoard['7'] == theBoard['8'] == theBoard['9'] != ' ': # across the top
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['7'] + " won. ****")                
                        break
                    elif theBoard['4'] == theBoard['5'] == theBoard['6'] != ' ': # across the middle
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['4'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ': # across the bottom
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['1'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['4'] == theBoard['7'] != ' ': # down the left side
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " + theBoard['1'] + " won. ****")
                        break
                    elif theBoard['2'] == theBoard['5'] == theBoard['8'] != ' ': # down the middle
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['2'] + " won. ****")
                        break
                    elif theBoard['3'] == theBoard['6'] == theBoard['9'] != ' ': # down the right side
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['3'] + " won. ****")
                        break 
                    elif theBoard['7'] == theBoard['5'] == theBoard['3'] != ' ': # diagonal
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['7'] + " won. ****")
                        break
                    elif theBoard['1'] == theBoard['5'] == theBoard['9'] != ' ': # diagonal
                        printBoard(theBoard)
                        print("\nGame Over.\n")                
                        print(" **** " +theBoard['1'] + " won. ****")
                        break 

                # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
                    if count == 9:
                        print("\nGame Over.\n")                
                        print("It's a Tie!!")
                        printBoard(theBoard)
                        break

                      
       

    # Now we will ask if player wants to restart the game or not.
    
    restart = input("Do want to play Again?(y/n)")
    if restart == "y" or restart == "Y":  
        
        for key in board_keys:
            theBoard[key] = " "
        

        game()

if __name__ == "__main__":
    game() 
```
